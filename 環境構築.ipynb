```python
keys = []
elems_th = browser.find_elements_by_tag_name('th')

for elem_th in elems_th:
    key = elem_th.text
    keys.append(key)
```

```python
keys
```

```python
values = []
elems_td = browser.find_elements_by_tag_name('td')
for elem_td in elems_td:
    value = elem_td.text
    values.append(value)
```

```python
values
```

```python
import pandas as pd
#からのdataframeを定義
df = pd.DataFrame()
df['項目'] = keys
df['値'] = values
```

```python
df
```

```python
df.to_csv('講師情報.csv',index=False)
```
