```python
import selenium
from selenium import webdriver
browser = webdriver.Chrome()
```

```python
#店舗詳細
details = []
names = []
categories = []
for page in range(1,79):
    browser.get('https://miyanotabetoku.jp/handling_store/result?type=search&page={}'.format(page))
    elems_stores_detail = browser.find_elements_by_class_name('index_item')
    #店舗詳細取得
    for elems_store_detail in elems_stores_detail:
        elems_store_detail = elems_store_detail.find_elements_by_class_name('store_detail')
        detail = []

        for elem_store_detail in elems_store_detail:
            store_detail = elem_store_detail.find_element_by_tag_name('dd').text
            detail.append(store_detail)

        #HPない場合　HPなしを追加
        if (detail[3] == 'あり' or detail[3] == 'なし'):
            detail.insert(3, 'なし')
        detail[1] = detail[1].split('\n')[0]
        #備考がない場合備考なしを追加
        if len(detail) < 7:
            detail.insert(6, '')
        details.append(detail)
        
    
    #店名取得
    stores_name = browser.find_elements_by_class_name('store_name')
    for store_name in stores_name:
        name = store_name.text
        names.append(name)
    
    #カテゴリー取得 index_item内のstore_categoryを取得
    stores_category = browser.find_elements_by_class_name('index_item')
    for store_category in stores_category:
        category = store_category.find_element_by_class_name('store_category').text
        categories.append(category)
```

```python
#ジャンルの取得
categories = []

stores_category = browser.find_elements_by_class_name('index_item')
for store_category in stores_category:
    category = store_category.find_element_by_class_name('store_category').text
    categories.append(category)
len(categories)
```

```python
# keyの取得
elems_dt = []
elems_dt = browser.find_elements_by_class_name('index_item')[0].find_elements_by_tag_name('dt')
keys = []
for elem_dt in elems_dt:
    key = elem_dt.text
    keys.append(key)

keys.insert(0, 'ジャンル')
keys.insert(0, '店名')
keys

```

```python
import pandas as pd
df = pd.DataFrame()
df['店名'] = names
df['カテゴリー'] = categories
df_details = pd.DataFrame(details)
df_details.columns= ['地域','住所','電話','ホームページ','テイクアウト','デリバリー','備考']
df_details
df = pd.concat([df,df_details], axis = 1)
df
df.to_csv('宮トク1209.csv', index = False)
```
