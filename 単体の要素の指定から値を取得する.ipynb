```python
from selenium import webdriver
browser = webdriver.Chrome()
browser.get('https://scraping-for-beginner.herokuapp.com/ranking/')
```

```python
###一つの観光地の情報取得
```

```python
values =[]
elems_h2 = browser.find_elements_by_tag_name('h2')
```

```python
for elem_h2 in elems_h2:
    value = elem_h2.text
    values.append(value)
    #数字の削除ができない
```
